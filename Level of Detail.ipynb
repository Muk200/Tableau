# **TYPES OF**

1. `include` : calc at a lower lvl of detail

2. `fixed` : specifiy exact LOD
3. `Exclude` : calc at a higher LOD

*syntax* :
{[in/ex/fi] field : what we want to do}

# **ADDDING LOD TO VIEW**
---
whether lod is aggregated or replicated in the view is determined by expression type & granularity(detail).

- `include` : Lod will have *same lvl* of detail as the view or *finer* lvl of detail.
        so, not replicated

- `fixed` : finer, coarser lvl of detail than view or same.
      needs to aggregate,depends on dim

- `exclude` : replicated vals appears , shows that the og is not actually aggregated & changing doesn't effect the view.

# **DIFFERENCE B/W TABLE CALC & LOD**
---

Here’s the content from the image in a tabular form:

| **Table Calculations**                                      | **LOD Expressions**                                                  |
|-------------------------------------------------------------|----------------------------------------------------------------------|
| Table Calculations are generated by Query Results.          | LOD Expressions are generated as part of a query to the data source. They depend on DBMS performance as they are expressed as a nested select. |
| Table Calculations can only produce results equal to or less granular than the said LOD. | LOD can produce results independent of the said LOD.                 |
| Dimensions that control the operations of a table are separate from the calculation syntax. | Dimensions that control the operations of an LOD expression are embedded in the expression itself. |
| Table Calculations are used as aggregated measures.         | LOD expressions can be used in other constructs.                     |
| Filters on Table Calculations act as a HIDE.                | Filters on LOD act as an EXCLUDE.                                    |
