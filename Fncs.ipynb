# String FNCS
---

allows to manipulate str data, like pull all the last names from all customer into a new field.

- `ASCII` : returns ascii code for first chr of str. ex- ascii(A)
- `char` : gives the chr encoded by ascii

- `find (str, substr, [start])` : gives index position of substr. Also it can have *start* param which makes it ignore every substr before that specified start pt

- `Find Nth (str,substr,occurrence)` : gives pos of Nth occurrence of substr.
- `Max` : max of (a,b).

  ex - ('apple', 'banana', 'banana')

- `Min` : gives min of (a,b)
- `left` : gives the Left *Most no of chrs*.
- `right` : rightMost not of chrs

- `mid (Str, start, [len])`: gives str starting at index pos from begining

- `space` : returns a str which has specified no of spaces.
ex - space (1) = one space
- `Ltrim` : gives str with any space removed from left.
- `Rtrim` : space removes from right
- `upper` : uppercase
- `lower` : gives with all chrs lowercase
- `contains` : returns true if str has specified substr.

  ex - contains('cal', 'alcu') = true
- `len` : gives the length of str
- `split(str, delimiter, token no)` : returns a substr from ip using a delimiter to divide it.
it is alt seq of delimiters & token
  ex - ('a-b-c-d', '-', 2) = b
- `Starts With` :  gives true if str starts with given substr
- `Ends With` : gives true if str ends with given substr
- `replace`
- `trim`: no spaces, Ltrim + Rtrim
